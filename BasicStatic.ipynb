```python
import pandas as pd
import numpy as np
```

```python
colnames = ['surgery', 'Age', 'rectal_t', 'pulse', 'resp_rate', 'temp_of_ex', 'pain', 'outcome']
df = pd.read_csv('https://raw.githubusercontent.com/obulygin/pyda_homeworks/master/statistics_basics/horse_data.csv',
                delimiter=',', na_values = '?', names = colnames, usecols = [0,1,3,4,5,6,10,22])
df.loc[(df.Age == 9), 'Age'] = 2
df = df.dropna(thresh = 6)
```

```python
def vb(date, col):
    q1 = date[col].quantile(0.25)
    q3 = date[col].quantile(0.75)   
    iqr = q3 - q1
    lower = q1 - (1.5 * iqr)
    higher = q3 + (1.5 * iqr)
    rem_out = date[date[col].between(lower, higher, inclusive = True)]
    return (pd.concat([date, rem_out]).drop_duplicates(keep=False)).dropna(subset = [col])
```

```python
df1 = vb(df, 'rectal_t')
df1
```

```python
df2 = vb(df, 'pulse')
df2
```

```python
df3 = vb(df, 'resp_rate')
df3
```

```python
df.loc[((df.outcome == 1) & (df.pain.isnull())), 'pain'] = 3
df.loc[((df['temp_of_ex'] == 2) & (df['rectal_t'].isnull())), 'rectal_t'] = 38.2
df = df[df['pulse'] < 145]
df = df[df['resp_rate'] <= 80]
df.loc[(((df['outcome'] == 2) | (df['outcome'] == 1)) 
        & (df['temp_of_ex'].isnull())), 'temp_of_ex'] = 1
df.loc[((((df['outcome'] == 2) | (df['outcome'] == 1)) 
         & (df['rectal_t'] > 38.0) & (df['rectal_t'] > 38.4)) & (df['temp_of_ex'].isnull())), 'temp_of_ex'] = 2
df.loc[(((df['Age'] == 1) & (df['outcome'] > 1.5)) & (df['temp_of_ex'].isnull())), 'temp_of_ex'] = 4
df["temp_of_ex"].fillna(3, inplace = True)
df['rectal_t'].fillna(38.16, inplace = True)
df.loc[((df['surgery'] == 1) & (df['temp_of_ex'] == 3) & (df['pain'].isnull())), 'pain'] = 5
df.loc[((df['surgery'] == 1) & (df['pain'].isnull())), 'pain'] = 4
df.loc[((df['surgery'] == 2) & (df['pulse'] < 52) & (df['pain'].isnull())), 'pain'] = 1
df['pain'].fillna(3, inplace = True)
df = df.dropna()
```
